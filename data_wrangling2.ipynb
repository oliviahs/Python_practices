```python
employee_actions = pd.read_csv('Employee_Actions.txt', sep='\t', encoding='latin1', dtype='unicode') #index_col=0,
employee_actions = employee_actions.iloc[:, :-1]
employee_actions.head()
```

```python
em_de = pd.read_csv('Employee_Demographics.txt', sep=",", names=['Post No','Index No','Organization','Gender','Nationality','Hire Date','Yrs Of Service','Age Group'])
em_de = em_de.iloc[1: , :]
```

```python
perf = pd.read_csv('Performance.txt', sep=",", names= ['Index No','Organization','Year','Performance Rating'])
perf = perf.iloc[1: , :]
perf.head()
```

```python
internal_app = pd.read_csv('Internal_Applicants.txt', sep=',',encoding='latin1', dtype='unicode', names=['Index No','Organization','Requisition No','System','Post No','Post title','Hired Index NO'])
internal_app = internal_app.iloc[1: , :]
internal_app.head()
```

```python
internal_app.shape
```

```python
vacancies = pd.read_csv('Summary_of_Vacancies.txt', sep=',', encoding='latin1', dtype='unicode', names=['Post No','Index No','Organization','Requisition No','Position Grade','No of Applications Submitted','Internal / External','Post title','Vacancy Year'
])
vacancies = vacancies.iloc[1:, :]
vacancies.head()
```

```python
# Joins
# First Left Join: 'Employee_Actions.txt' <-- 'Employee_Demographics.txt'
```

```python
first_join = pd.merge(left=employee_actions, right=em_de, how='left', on = ['Post No','Index No'])
first_join.head()
```

```python
first_join.shape #296000
```

```python
# Second Left Join: 'Employee_Actions.txt' <-- 'Employee_Demographics.txt' <-- 'Internal_Applicants.txt'
# 한 명이 이직을 위해 여러군데 여러번 지원했을수있으니까 숫자가 엄청 늘어난다. 
second_join = pd.merge(left=first_join, right=internal_app, how = 'left', on=['Index No'])
second_join.head()
```

```python
second_join.shape
```

```python
# Third Left Join : 'Employee_Actions.txt' <-- 'Employee_Demographics.txt' <-- 'Internal_Applicants.txt' <-- 'Summary_of_Vacancies.txt'

third_join = pd.merge(left = second_join, right= vacancies, how='left', left_on=['Post No_x', 'Index No'], right_on=['Post No','Index No'])
third_join.shape
```

```python
# Join with 'Location.xlsx' led to too many rows...(64195199, 36)
```

```python
# Fourth Left Join : 'Employee_Actions.txt' <-- 'Employee_Demographics.txt' <-- 'Internal_Applicants.txt' 
# <-- 'Summary_of_Vacancies.txt' <-- 'Performance.txt'
join4 = pd.merge(left = third_join, right = perf, how='left', on=['Index No'])

join4.shape
```
